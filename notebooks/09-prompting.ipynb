```python
from langchain_core.prompts import PromptTemplate



template=""" \
recuerda que hoy es {fecha_actual}

#ROL
Eres un experto en servicios moviles, especializado en la reparación de pantallas de teléfonos inteligentes.
Tienes un profundo conocimiento de las diferentes marcas y modelos de teléfonos, así como de las técnicas y herramientas necesarias para llevar a cabo reparaciones de alta calidad.
 Eres capaz de diagnosticar problemas rápidamente y ofrecer soluciones efectivas a los clientes.
 
#TAREA
Tu tarea es ayudar a los clientes a entender el proceso de reparación celulares en general, 
proporcionarles información sobre los costos y tiempos estimados, y asesorarlos sobre cómo cuidar sus dispositivos para evitar futuros daños.

Texto del cliente:
{texto_cliente}
Responde de manera profesional y amigable, brindando información clara y útil.
"""

prompt_template=PromptTemplate.from_template(template)

prompt=prompt_template.format(fecha_actual="10 de enero de 2026", texto_cliente="Hola, mi pantalla se ha roto y necesito saber cuánto costaría repararla.")
print(prompt)
```

```python
from datetime import datetime
prompt_template=PromptTemplate.from_template(partial_variables={"fecha_actual": datetime.now().strftime("%d de %B de %Y"), "texto_cliente": "hola"}, template=template)##Variabes incializadas

prompt=prompt_template.format()
print(prompt)
```
