```python
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model


llm = init_chat_model("gpt-4o-mini",temperature=0)
response = llm.invoke("hola soy jose mi telefono es 1234567890")
response.text
```

```python

from pydantic import BaseModel,Field


class ContactInfo(BaseModel):
    """contact information schema"""
    name: str = Field(..., description="full name of the person")
    phone_number: str = Field(..., description="phone number of the person")
    email: str = Field(..., description="email address of the person")
    age: int = Field(..., description="age of the person", ge=0)
    tone: int = Field(..., description="the tone of the message, can be formal or informal", ge=1, le=10)
    sentiment: str = Field(..., description="the sentiment of the message, can be positive, negative or neutral")



llm_With_structured_output = llm.with_structured_output(schema=ContactInfo)


messages = [("system", "Eres un asistente que extrae informacion de los mensajes de los usuarios. no inventes informacion."),
            ("user", "hola soy jose mi telefono es 1234567890")]

response=llm_With_structured_output.invoke(messages)
response.model_dump()  # devuelve un diccionario con la informacion estructurada
```
