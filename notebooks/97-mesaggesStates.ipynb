```python
from langchain_core.messages import SystemMessage

def system_node(state):
    return {
        "messages": [
            SystemMessage(content="Eres un asistente de prueba.")
        ]
    }
print({"messages": [ SystemMessage(content="Eres un asistente de prueba.")]})
```

```python
from langchain_core.messages import HumanMessage

def user_node(state):
    return {
        "messages": [
            HumanMessage(content=f"Hola, soy {state['user_name']}")
        ]
    }
print({"messages": [ HumanMessage(content="Hola, soy Alice")]})
```

```python
from langchain_core.messages import AIMessage

def assistant_node(state):
    ia_mesagge=[
            AIMessage(content="Encantado de conocerte ðŸ˜Š")
        ]
    return {
        "messages": ia_mesagge
    }
print({"messages": [ AIMessage(content="Encantado de conocerte ðŸ˜Š")]})
```

```python
def aggregator(state):
    print("\nðŸ§  MENSAJES FINALES:\n")
    for msg in state["messages"]:
        print(f"{msg.type.upper()}: {msg.content}")

    return {}  # no modificamos nada
```

```python
from langgraph.graph import StateGraph, START, END

builder = StateGraph(State)

builder.add_node("system", system_node)
builder.add_node("user", user_node)
builder.add_node("assistant", assistant_node)
builder.add_node("aggregator", aggregator)

builder.add_edge(START, "system")
builder.add_edge("system", "user")
builder.add_edge("user", "assistant")
builder.add_edge("assistant", "aggregator")
builder.add_edge("aggregator", END)

agent = builder.compile()
```

```python
initial_state = {
    "user_name": "Juan"
}

agent.invoke(initial_state)
```
